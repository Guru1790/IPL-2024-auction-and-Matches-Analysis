```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the  dataset
auction_df = pd.read_csv('/content/auction.csv')
matches_df = pd.read_csv('/content/ipl2024 Matches.csv')
```

```python
# Display the first few rows of each dataset
print("Auction Dataset:")
auction_df.head()
```

```python
print("\nMatches Dataset:")
matches_df.head()
```

```python
print("\nMissing Values in Matches Dataset:")
print(matches_df.isnull().sum())
```

```python
# Display data types
print("\nData Types in Auction Dataset:")
print(auction_df.dtypes)
```

```python
print("\nData Types in Matches Dataset:")
print(matches_df.dtypes)
```

```python
matches_df['date'] = pd.to_datetime(matches_df['date'], format='%B %d,%Y')
```

```python
print("\nAuction Dataset Info:")
print(auction_df.info())
```

```python
print("\nMatches Dataset Info:")
print(matches_df.info())
```

```python
# Summary statistics
print("\nSummary Statistics of Auction Dataset:")
print(auction_df.describe())
```

```python
print("\nSummary Statistics of Matches Dataset:")
print(matches_df.describe())
```

```python
# Distribution of players by nationality
print("\nDistribution of Players by Nationality:")
print(auction_df['NATIONALITY'].value_counts())
```

```python
# Distribution of players by team
print("\nDistribution of Players by Team:")
print(auction_df['TEAM'].value_counts())
```

```python
# Number of matches won by each team
print("\nNumber of Matches Won by Each Team:")
print(matches_df['winner'].value_counts())
```

```python
# Top 10 players by price paid
print("\nTop 10 Players by Price Paid:")
print(auction_df.nlargest(10, 'PRICE PAID')[['PLAYER', 'TEAM', 'PRICE PAID']])
```

```python
# Winning percentage of teams based on toss decision
toss_decision_outcome = matches_df.groupby(['toss_winner', 'decision']).agg({'winner': 'count'}).reset_index()
toss_decision_outcome['win_percentage'] = (toss_decision_outcome['winner'] / toss_decision_outcome['winner'].sum()) * 100
```

```python
# Top 10 highest scoring matches (Second Innings)
print("\nTop 10 Highest Scoring Matches (Second Innings):")
print(matches_df.nlargest(10, 'second_score')[['team1', 'team2', 'second_score']])
```

```python
# Top 10 highest scoring matches
print("\nTop 10 Highest Scoring Matches (First Innings):")
print(matches_df.nlargest(10, 'first_score')[['team1', 'team2', 'first_score']])
```

```python
# Most valuable players (based on player_of_the_match awards)
print("\nMost Valuable Players (Player of the Match Awards):")
print(matches_df['player_of_the_match'].value_counts().head(10))
```

```python
# Group matches by teams and summarize
team_performance = matches_df.groupby('winner').agg({
    'id': 'count',
    'first_score': 'sum',
    'second_score': 'sum',
    'most_runs': 'sum',
    'most_wkts': 'sum'
}).reset_index().rename(columns={'id': 'matches_won'})
```

```python
# Merge with auction data
team_auction_summary = auction_df.groupby('TEAM').agg({
    'PRICE PAID': 'sum'
}).reset_index()
```

```python
# Merge team performance with auction summary
merged_df = pd.merge(team_performance, team_auction_summary, left_on='winner', right_on='TEAM', how='left')
print("\nMerged DataFrame:")
print(merged_df.head())
```

```python
# Correlation analysis
numeric_columns = merged_df.select_dtypes(include=['number']).columns
print("\nCorrelation Matrix:")
print(merged_df[numeric_columns].corr())
```

```python
# Winning percentage of teams based on toss decision
toss_decision_outcome = matches_df.groupby(['toss_winner', 'decision']).agg({'winner': 'count'}).reset_index()
toss_decision_outcome['win_percentage'] = (toss_decision_outcome['winner'] / toss_decision_outcome['winner'].sum()) * 100

plt.figure(figsize=(14, 8))
sns.barplot(data=toss_decision_outcome, x='toss_winner', y='win_percentage', hue='decision')
plt.title('Winning Percentage of Teams Based on Toss Decision')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of match outcomes by decision (bat/field)
plt.figure(figsize=(10, 6))
sns.countplot(data=matches_df, x='decision')
plt.title('Distribution of Match Outcomes by Decision')
plt.show()
```

```python
# Distribution of prices paid for players by team
plt.figure(figsize=(14, 8))
sns.boxplot(data=auction_df, x='TEAM', y='PRICE PAID')
plt.title('Distribution of Prices Paid for Players by Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Matches won by team
plt.figure(figsize=(14, 8))
sns.countplot(data=matches_df, x='winner', order=matches_df['winner'].value_counts().index)
plt.title('Number of Matches Won by Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram of Player Prices
plt.figure(figsize=(10, 6))
sns.histplot(auction_df['PRICE PAID'], bins=20, kde=True)
plt.title('Distribution of Player Prices')
plt.xlabel('Price Paid')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Plot of Most Runs by Player
top_runs = matches_df.groupby('most_runs')['most_runs'].count().nlargest(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_runs.index, y=top_runs.values)
plt.title('Top 10 Players with Most Runs')
plt.xlabel('Player')
plt.ylabel('Number of Matches with Most Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar Plot of Most Wickets by Player
top_wickets = matches_df.groupby('most_wkts')['most_wkts'].count().nlargest(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_wickets.index, y=top_wickets.values)
plt.title('Top 10 Players with Most Wickets')
plt.xlabel('Player')
plt.ylabel('Number of Matches with Most Wickets')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot of Team Performance Over Time
team_performance_over_time = matches_df.groupby(['date', 'winner']).size().unstack(fill_value=0)
team_performance_over_time = team_performance_over_time.cumsum()

plt.figure(figsize=(14, 8))
for team in team_performance_over_time.columns:
    plt.plot(team_performance_over_time.index, team_performance_over_time[team], label=team)
plt.title('Team Performance Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Wins')
plt.legend()
plt.show()
```

```python
# Heatmap of Correlation Matrix
# Convert potential mixed-type columns to numeric, coercing errors to NaN
for column in merged_df.columns:
    merged_df[column] = pd.to_numeric(merged_df[column], errors='coerce')
```

```python
# Drop columns with all NaN values (non-numeric)
merged_df = merged_df.dropna(axis=1, how='all')

plt.figure(figsize=(12, 8))
sns.heatmap(merged_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Violin Plot of Player Prices by Team
plt.figure(figsize=(14, 8))
sns.violinplot(data=auction_df, x='TEAM', y='PRICE PAID')
plt.title('Distribution of Player Prices by Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count Plot of Matches Played per Month
matches_df['month'] = matches_df['date'].dt.month_name()
plt.figure(figsize=(12, 8))
sns.countplot(data=matches_df, x='month', order=matches_df['month'].value_counts().index)
plt.title('Number of Matches Played per Month')
plt.xlabel('Month')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
# Joint Plot of Price Paid vs Most Runs
# First, merge player price with match performance
performance_df = matches_df[['most_runs']].copy()
performance_df = performance_df.merge(auction_df[['PLAYER', 'PRICE PAID']], left_on='most_runs', right_on='PLAYER', how='left')

plt.figure(figsize=(10, 6))
sns.jointplot(data=performance_df, x='PRICE PAID', y='most_runs', kind='scatter', height=8)
plt.suptitle('Price Paid vs. Most Runs', y=1.02)
plt.xlabel('Price Paid')
plt.ylabel('Most Runs')
plt.show()
```

```python
# Pair Plot of Key Statistics
# We select key columns from matches_df and auction_df for the pair plot
key_stats = matches_df[['first_score', 'second_score', 'most_runs', 'most_wkts']].copy()
key_stats['PRICE PAID'] = matches_df['most_runs'].map(auction_df.set_index('PLAYER')['PRICE PAID'])

sns.pairplot(key_stats)
plt.suptitle('Pair Plot of Key Statistics', y=1.02)
plt.show()
```

```python
# Swarm Plot of Toss Decisions by Winner
plt.figure(figsize=(14, 8))
sns.swarmplot(data=matches_df, x='toss_winner', y='id', hue='decision')
plt.title('Toss Decisions by Winning Team')
plt.xlabel('Team')
plt.ylabel('Match ID')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of matches won by each team
win_counts = matches_df['winner'].value_counts()

# Count total matches played by each team
total_matches = pd.concat([matches_df['team1'], matches_df['team2']]).value_counts()

# Calculate win rates
win_rates = win_counts / total_matches * 100

# Convert to DataFrame for easier plotting
win_rates_df = pd.DataFrame({'Team': win_rates.index, 'Win Rate (%)': win_rates.values})

# Bar Plot of Win Rates
plt.figure(figsize=(12, 8))
sns.barplot(data=win_rates_df, x='Team', y='Win Rate (%)', palette='viridis')
plt.title('Win Rate of Each Team (%)')
plt.xlabel('Team')
plt.ylabel('Win Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
